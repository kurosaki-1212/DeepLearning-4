```python
# 人間椅子のテキストを300行程度読み込んで、
# 使われている品詞のそれぞれの個数を出力しましょう。
# また、出力頻度順（出現回数）に並び替えて、「品詞（出現回数）」として表示しましょう。

import MeCab
text = ''

with open('ningen_isu.txt', 'r', encoding='shift-jis') as f:
    text = f.read()

tagger = MeCab.Tagger('/usr/lib/x86_64-linux-gnu/mecab/dic/mecab-ipadic-neologd')
# MeCabの形態素解析を行う
# 最初のnodeを取得
node =  tagger.parseToNode(text)

# hinshi_dictは、{'品詞':'出現回数'}の形式でデータを保持する辞書型
hinshi_dict = {}

# nodeに入っているデータ分ループを回す
while node:
    # 形態素解析を行った１要素から品詞部分だけ取り出す
    hinshi = node.feature.split(',')[0]
    # hinshi_dictに存在する品詞か？
    if hinshi in hinshi_dict.keys():
        hinshi_dict[hinshi] +=  1
    else:
        hinshi_dict[hinshi] = 1
    node = node.next

# hinshi_dictのvalue値を並び替える
hinshi_dict = sorted(hinshi_dict.items(), key = lambda x:x[1], reverse = True)
# 並び替えたデータを表示
for key in hinshi_dict:
    print('{}:{}回'.format(key[0], key[1]))
```
