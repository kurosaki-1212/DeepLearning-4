3.  judge = lambda a, b: (a - b + 3) % 3 が何をやっているのか自分の言葉で説明できるようにコメントに記載する。以下の内容を網羅すること。

> ①引数には、どんな値が設定されるか？
> ②戻り値には、どんな値を設定するか？
> ヒント：lambda 引数:戻り値

4.  「janken-data.pkl」にはどのようなデータが保存されているか、説明できるようにコメントに記載する。


```python
# ジャンケンデータを作成する
import pickle
import numpy as np

# ジャンケンの手と結果を定義
hands = {'グー':0, 'チョキ':1, 'パー':2}
results = ['あいこ', '負け', '勝ち']

# じゃんけんの公式を定義
judge = lambda a, b: (a - b + 3) % 3

# ランダムにデータを作成
import random
random_hand = lambda : random.randint(0, 2)

x_items = []
y_items = []
for i in range(3000):
  a = random_hand()
  b = random_hand()
  result = judge(a, b)
  x_items.append([a, b])
  y_items.append(result)


# 作成したデータを表示
print(x_items)
print(y_items)

# データを学習用とテスト用に分割
x_train = x_items[0:2000]
y_train = y_items[0:2000]
x_test = x_items[2000:]
y_test = y_items[2000:]

# データを保存
items = [[x_train, y_train],
         [x_test, y_test]]
with open('janken-data.pkl', 'wb') as fp:
  pickle.dump(items, fp)
```

2.   tf.keras.layers.Dense(30, activation='relu', input_dim=2) は、中間層の活性化関数に「ReLu関数」を使用することを表し、入力層が2、中間層の数が30であることを表す。「ReLu」関数とは何か。どんな特徴を持つ関数か。 - 活性化関数のうちの一つで0以上ならもらった値をそのまま返す。また勾配消失などにも強い

8.   この学習は、回帰、分類どちらの学習を行っているか。 - 分類


```python
# テキストP258（janken-train.py）の内容の実装
import tensorflow as tf
import numpy as np
import pickle

# 保存したじゃんけんのデータを読み込む
with open('janken-data.pkl', 'rb') as fp:
  data = pickle.load(fp)
(x_train, y_train), (x_test, y_test) = data

# 学習モデルを構築
# Sequential→逐次実行
model = tf.keras.models.Sequential([
# Dense→全結合ニューラルネットワーク
# tf.keras.layers.Dense(ユニット数(中間層の数), 活性化関数,入力層の数)
  tf.keras.layers.Dense(30, activation = 'relu', input_dim = 2),
# tf.keras.layers.Dense(ユニット数(出力層の数), 活性化関数)
  tf.keras.layers.Dense(3, activation = 'softmax')
])
model.compile(optimizer = 'adam',
    # 損失関数(交差エントロピー誤差) categorical_crossentropy は通常 One-Hotになっている
    # sparse→ラベルエンコーディング
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy'])

# 学習
model.fit(x_train, y_train, epochs = 20)
# テストデータを評価(evaluate)
model.evaluate(x_test, y_test, verbose = 2)

# 実際に勝負
def janken(a, b):
  hands = {'グー':0, 'チョキ':1, 'パー':2}
  results = ['あいこ', '負け', '勝ち']
  x = np.array([[hands[a], hands[b]]])
  r = model.predict(x)
  print(r)
  print(a, b, '→', results[r[0].argmax()])

janken('グー', 'グー')
janken('チョキ', 'パー')
janken('パー', 'チョキ')
```
