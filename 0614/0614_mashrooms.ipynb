```python
import tensorflow as tf
import tensorflow.keras as keras
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# きのこデータの読み込み
mushrooms_data = pd.read_csv("mushrooms.csv")

# class列のみラベルエンコーディング
le = LabelEncoder()
encoded = le.fit_transform(mushrooms_data['class'].values)
mushrooms_data['class'] = encoded

# 説明変数と目的変数に分割する
y = mushrooms_data['class']
x = mushrooms_data.drop('class', axis = 1)

# 説明変数をダミー変数化する
x = pd.get_dummies(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

# モデル構造を定義
Dense = keras.layers.Dense
model = keras.models.Sequential()
model.add(Dense(10, activation='relu', input_shape=(117,)))
model.add(Dense(2, activation='softmax'))

# モデルを構築
model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy'])

# 学習を実行
model.fit(x_train, y_train,
    batch_size=20,
    epochs=300)

# モデルを評価 
score = model.evaluate(x_test, y_test, verbose=1)
print('正解率=', score[1], 'loss=', score[0])
```
