```python
%matplotlib inline
from sklearn.model_selection import train_test_split
import keras

# dataset読み込み用
from sklearn import datasets

# fetch_olivetti_facesデータセットを読み込む
faces = datasets.fetch_olivetti_faces()

# データの形状を確認（400人の顔データ／64*64=4096ピクセルデータという意味）
faces.data.shape
```

```python
#説明変数と目的変数に分離する
# 説明変数 = x
x = faces.data
# 目的変数 = y
y = faces.target
```

```python
# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# 説明変数データを64*64=4096 の一次元配列に変換し正規化（255で割る）
x_train = x.reshape(-1, 4096).astype('float32') / 255
x_test = x.reshape(-1, 4096).astype('float32') / 255
```

```python
# 目的変数データのOneHotエンコーディング
from tensorflow.keras import utils as np_utils

y_train = np_utils.to_categorical(y.astype('int32'))
y_test = np_utils.to_categorical(y.astype('int32'))
```

```python
in_size = 64 * 64
out_size = 40

# モデル構造を定義
Dense = keras.layers.Dense
model = keras.models.Sequential()
model.add(Dense(512, activation='relu', input_shape=(in_size,)))
model.add(Dense(out_size, activation='softmax')) 

# モデルを構築
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy'])

# 学習を実行
model.fit(x_train, y_train,batch_size=20,epochs=60)

# モデルを評価
score = model.evaluate(x_test, y_test, verbose=1)
print('正解率=', score[1], 'loss=', score[0])
```
