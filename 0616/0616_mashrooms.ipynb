```python
import numpy as np
import pandas as pd
import sklearn

# データの読み込み
df = pd.read_csv('mushrooms.csv')

# キノコの様子を確認
df.head()
```

```python
# class列は正解データなので、poisonous→1（食べられない） edible→0（食べられる）として使うことにする。
# ラベルエンコーダを使う
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
# class列のみラベルエンコーディング
encoded = le.fit_transform(df['class'].values)
df['encoded'] = encoded
# できてるか確認
df.head()
```

```python
# エンコーディングの成功が確認できたため、class列をdropし、encoded列を正解データとして扱う
df1 = df.iloc[:,1:]
df1.head()
```

```python
# 最後の encoded 以外は全部ダミー変数化する
df_dummy = pd.get_dummies(df1.iloc[:,1:], )
```

```python
# 様子を確認
df_dummy.head()
```

```python
# 特徴量
x = df_dummy.iloc[:,1:]
x.head()
```

```python
# 正解
y = df_dummy.iloc[:,:1]
y.head()
```

```python
# 学習用とテスト用に分割する
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
```

# ここからディープラーニングの実装を行う

```python
# モデル構造を定義
Dense = keras.layers.Dense
model = keras.models.Sequential()
model.add(Dense(30, activation = 'relu', input_dim = 111))
model.add(Dense(2, activation = 'softmax')) 

# モデルを構築
model.compile(
    loss = 'sparse_categorical_crossentropy',
    optimizer = 'adam', 
    metrics=['accuracy'])

# 学習を実行*
model.fit(x_train, y_train,batch_size=20,epochs=20)

# モデルを評価 
score = model.evaluate(x_test, y_test, verbose=1)
print('正解率=', score[1], 'loss=', score[0])
```
